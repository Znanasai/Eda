## Empirical Rule

```python
One Standard Deviation (Âµ Â± Ïƒ): 68% of Data

Two Standard Deviation (Âµ Â± 2Ïƒ): 95% of Data

Three Standard Deviation (Âµ Â± 3Ïƒ): 99.7% of Data
```

```python
mu=rt['prevailing_wage'].mean()
sigma=rt['prevailing_wage'].std()*2
lb=mu-sigma
ub=mu+sigma
con1=(rt['prevailing_wage']<ub)&(rt['prevailing_wage']>lb)
rt[con1]
```

```python
import numpy as np
import matplotlib.pyplot as plt
q1 = np.percentile(wage_data, 25)
q2 = np.percentile(wage_data, 50)  # Median
q3 = np.percentile(wage_data, 75)
IQR = q3 - q1
lb = q1 - 1.5 * IQR
ub = q3 + 1.5 * IQR
con1 = wage_data < lb
con2 = wage_data > ub
con3 = con1 | con2
wage_data[con3] = q2
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.boxplot(wage_data,vert=False)
plt.title("Boxplot of Wage Data after Outlier Replacement")
plt.subplot(1, 2, 2)
plt.hist(wage_data, bins=30, edgecolor='black')
plt.title("Histogram of Wage Data after Outlier Replacement")
plt.show()
```

```python
con1=rt['continent']=='Asia'
con2=rt['case_status']=='Certified'
con3=con1&con2
rt[con3]
```

```python
continents=rt['continent'].unique()
certified_list,denied_list=[],[]
for i in continents:
    con1=rt['continent']==i
    con2=rt['case_status']=='Certified'
    con4=rt['case_status']=='Denied'
    con3=con1&con2
    den_c=con1&con4
    print(f'the count of case status accepted is {len(rt[con3])}')
    print(f'the count of case status denied is {len(rt[den_c])}')
    certified_list.append(len(rt[con3]))
    denied_list.append(len(rt[den_c]))
```

```python
cols=['Certified','Denied']
df4=pd.DataFrame(zip(certified_list,denied_list),index=continents,columns=cols)
df4
```
