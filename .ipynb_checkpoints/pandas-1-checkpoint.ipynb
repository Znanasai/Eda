```python
continents=[]
no_of_Applicants=[]
for i in rt['continent'].unique():
    con=rt['continent']==i
    count=len(rt[con])
    continents.append(i)
    no_of_Applicants.append(count)
zip(continents,no_of_Applicants)
cols=['continents','no_of_Applicants']
right_df=pd.DataFrame(zip(continents,no_of_Applicants),columns=cols)
right_df.to_csv('Continent_visa_details.csv',index=False)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
fs=rt['continent'].value_counts()
continents=fs.keys()
no_of_apps=fs.values
cols=['continents','no_of_apps']
se=pd.DataFrame(zip(continents,no_of_apps),columns=cols)
plt.bar('continents','no_of_apps',data=se)
plt.title('PLOT')
plt.xlabel('Continents')
plt.ylabel('Number of applicants')
```

```python
for i in rt.columns[1:]:
    fs=rt[i].value_counts()
    continents=fs.keys()
    no_of_apps=fs.values
    cols=[i,'no_of_apps']
    se=pd.DataFrame(zip(i,no_of_apps),columns=cols)
    plt.bar(i,'no_of_apps',data=se)
    plt.title('PLOT')
    plt.xlabel(i)
    plt.ylabel('Number of applicants')
    path=r"C:\Users\dell\online_cls\EDA\df1"
    plt.savefig(f'{path}\\Data_{i}.jpg')
```
